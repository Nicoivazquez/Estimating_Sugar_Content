# Part 2: Parsing Quickly

### I had to manually stem some of the words for better performance.

```python
import re
cats = ['Pineapple',
        'Apple',
        'Banana',
        'Grape',
        'Peach', 
        'nectari', 
        'Strawber',
        'Blueberr',
        'berr',
        'Citrus',
        'Melon',
        'Driedfruit', 
        'raisin',
        'Pear',
        'Mango',
        'papaya']

dicts = {'Apple': 1,
        'Banana':2,
        'Grape':3,
        'Peach':4, 'nectari':4, 
        'Strawber':5,
        'Blueberr': 6, 'berr':6,
        'Citrus':7,
        'Melon':8,
        'Driedfruit':9, 'raisin':9,
        'Pear':10,
        'Pineapple' :11,
        'Mango': 12,
        'papaya' : 12}

def preprocess(text):
    clean_data = ''
    for x in (text[:]): 
        new_text = re.sub('<.*?>', '', x)   # remove HTML tags
        new_text = re.sub(r'[^\w\s]', '', new_text) # remove punctuation
        new_text = re.sub(r'\d+','',new_text)# remove numbers
        new_text = new_text.lower() # lower case  
        if new_text != '':
            clean_data+= new_text
    return clean_data

def my_func(x,cats=cats,dicts=dicts):
    for cat in cats:
        if cat.lower() in preprocess(x):
            return dicts[cat]
    else:
        return 0

df['cat?'] = df["Main food description"].apply(my_func)
```

```python
df['cat?'].value_counts() # checking the value counts making sure we got a food for each category.
```

# Part 3: Finding the average for each category

```python
avg_apple = df[df['cat?'] == 1]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 1]) # finding the average sugar for each
avg_apple
```

```python
avg_apple = df[df['cat?'] == 1]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 1])
avg_Banana = df[df['cat?'] == 2]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 2])
avg_Grape = df[df['cat?'] == 3]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 3])
avg_Peach_nectarine = df[df['cat?'] == 4]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 4])
avg_Strawberies = df[df['cat?'] == 5]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 5])
avg_Blueberries_berries = df[df['cat?'] == 6]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 6])
avg_Citrus = df[df['cat?'] == 7]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 7])
avg_Melon = df[df['cat?'] == 8]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 8])
avg_Driedfruit = df[df['cat?'] == 9]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 9])
avg_pear = df[df['cat?'] == 10]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 10])
avg_Pineapple = df[df['cat?'] == 11]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 11])
avg_Mango_papaya = df[df['cat?'] == 12]['Sugars, total\n(g)'].sum()/len(df[df['cat?'] == 12])
```

# Part 4: Sorting values and printing category names

```python
fruits_sorted = np.argsort([avg_Banana, avg_apple, avg_Blueberries_berries, avg_Citrus, avg_Driedfruit, avg_Grape, avg_Mango_papaya, avg_Melon, avg_Peach_nectarine, avg_pear, avg_Pineapple, avg_Strawberies])
fruits_sorted = fruits_sorted +1 # adding one to the values for printing key dictionary pair below
```

```python
def get_key(val):
    for key, value in dicts.items():
         if val == value:
            return key
        
for value in fruits_sorted:
    print(get_key(value))
```
